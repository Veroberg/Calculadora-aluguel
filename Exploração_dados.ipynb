```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
URL = "https://www.vivareal.com.br/aluguel/sp/sao-paulo/?pagina=7"
```

```python
imoveis = pd.read_csv(url)
```

```python
imoveis.head()
```

```python
imoveis.info()
```

```python
# Estatística descritiva

imoveis.describe().T 
```

```python
imoveis[['aluguel', 'area']]
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(imoveis[['aluguel', 'area']])
plt.ylabel('Aluguel')
plt.xlabel('Area')
plt.title('Gráfico');
```

```python
imoveis['area'].values
```

```python
plt.plot(imoveis['area'].values);
```

```python
# O que influencia o aumento dos aluguéis?

plt.figure(figsize=(15, 5))
sns.barplot(x='aluguel', y='area', data=imoveis, hue='quarto');
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(y='area', x='aluguel', data=imoveis[:20], hue='quarto');
```

***Dois dormitórios são os mais presentes na busca de uma imóvel para alugar. E quanto maior a área do imóvel, maios o valor do aluguél.***


```python
plt.figure(figsize=(15, 5))
sns.barplot(y='area', x='aluguel', data=imoveis[:20], hue='quarto');
```
